What is Pandas?

Pandas is a Python library used for working with data. It helps you load, clean, organize, analyze, and manipulate data easily.

Think of Pandas like Excel inside Python, but much more powerful.

For example:

Excel	                  Pandas
Rows & Columns	          Rows & Columns
Filter data	              Filter data
Sort	                  Sort
Formulas	              Calculations
Pivot Table 	          Grouping/Aggregation
Cleaning duplicates	      Data cleaning

Why do Data Analysts use Pandas?

As a data analyst, most of your work is:

Collecting data
Cleaning messy data
Analyzing data
Finding insights
Making reports

1. Series in Pandas

A Series is a single column of data.

Think of it like one column in Excel.

Example:

```python
s = pd.Series([10, 20, 30, 40])

s
```

Easy Understanding:

Imagine marks of students:

Marks
10
20
30
40

This single column = Series

Remember:
Series = One Column

2. DataFrame in Pandas

A DataFrame is a table with rows and columns.

Think of it like an Excel sheet.

Example:

```python
data = {
    "Name": ["Ram", "Sam", "John"],
    "Age": [22, 24, 21]
}

df = pd.DataFrame(data)

df
```

Easy Understanding:
Name	Age
Ram	    22
Sam	    24
John	21

This whole table = DataFrame

Remember:
DataFrame = Full Table

3. Add Another Column

Suppose we want to add Marks column.

Existing Data:

Name	Age
Ram	    22
Sam	    24

Add new column:

```python
df["Marks"] = [80, 90, 100]

df
```

Easy Formula: df["New Column"] = values

Example:

df["City"] = ["Chennai", "Coimbatore"]
