3. Outliers
What are outliers?

Unusual/extreme values that are very different.

Example salary:

30000
35000
40000
45000
1000000 ← OUTLIER

See?

Everyone salary is normal but one is too high.

That unusual value = outlier

Outliers can affect:

average
machine learning
analysis

How to find outliers?

Using box plot.

5. Skewness (Skew)
What is skewness?

It tells:

Is data leaning left or right?

Think of it like data shape.

1. Symmetrical (Normal)

Balanced both sides.

Skew ≈ 0

2. Positive Skew (Right Skew)

Long tail right side.

Skew > 0

Example:
Salary data usually.

Few people earn very high salary.

3. Negative Skew (Left Skew)

Long tail left side.

Skew < 0

Easy Rule
Skew Value	Meaning
0	  Normal
> 0   Right skew
< 0	  Left skew

```python
data["Current_Salary"].skew()
```

```python
data["Employee_age"].skew()
```

6. Kurtosis
What is kurtosis?

Kurtosis tells:

How many extreme values (outliers) exist?

OR

How sharp or flat data is

High kurtosis

Many outliers.

Sharp peak.

Low kurtosis

Few outliers.

Flat shape.

```python
data["Current_Salary"].kurt()
```

```python
data["years_experience"].kurt()
```

Easy Rule
Value	Meaning
> 3	    More outliers
< 3	    Fewer outliers

(For now just remember this simple rule.)
