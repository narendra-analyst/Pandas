4. Remove Null Values
What is Null?

Null means missing value.

Example:

Name	Marks
Ram	    80
Sam	    NaN
John	90

Here Sam's marks are missing.

Remove null rows

df.drop_duplicates()

```python
data = {
    "Name": ["Ram", "Sam", "John"],
    "Marks": [80, None, 90]
}

df = pd.DataFrame(data)

df.dropna()
```

```python
data = {
    "Name": ["Ram", "Sam", "Ram"],
    "Marks": [80, 90, 80]
}

df = pd.DataFrame(data)

df.drop_duplicates()
```

Duplicate row removed.

Easy Remember:

What is fillna() in Pandas?

fillna() is used to fill missing (null/NaN) values instead of removing them.

Remember:

dropna() → removes rows with null values 
fillna() → fills null values 

Why do we use fillna()?

Suppose you have this data:

Name	Marks
Ram	    80
Sam	    NaN
John	90

Sam's marks are missing.

If we use: df.dropna()

Sam's entire row gets deleted.

But sometimes we don't want to lose data.

So we use:fillna()

to fill the missing value.

Syntax
df.fillna(value)

value = what you want to replace the missing value with.

Example 1: Fill with 0

```python
data = {
    "Name": ["Ram", "Sam", "John"],
    "Marks": [80, None, 90]
}

df = pd.DataFrame(data)

df.fillna(0)
```

Easy understanding:

Missing marks became 0.

Example 2: Fill with a specific value

```python
df.fillna(50)
```

Example 3: Fill only one column

Suppose:

Name	Marks
Ram	    80
Sam	    NaN

Fill only Marks column:

```python
data = {
    "Name": ["Ram", "Sam"],
    "Marks": [80, None]
}

df = pd.DataFrame(data)

df.fillna(0)
```

```python
df["Marks"].fillna(0)
```

Only marks column changed.

But dataframe stays same.

To save permanently:

Method 1 (Recommended

df = df.fillna(0)

Method 2

```python
df.fillna(0, inplace=True)
```

Now changes are saved in df.

Real-life Data Analyst Example

Imagine salary data:

Name	Salary
A	    50000
B	    NaN
C	    60000

Instead of deleting employee B:

```python
data = {
    "Name": ["A", "B", "C"],
    "Salary": [50000, 'NaN', 60000]
}

df = pd.DataFrame(data)

df.fillna(0)
```

```python
df["Salary"] = df["Salary"].fillna(50000)
```

You fill missing salary with average/default value.

This is common in data cleaning.
